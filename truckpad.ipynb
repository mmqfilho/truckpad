### Análise de dados

```python
import pandas as pd
import MySQLdb
import pandas.io.sql as psql

%matplotlib notebook

db=MySQLdb.connect(host='localhost', user='root', passwd='', db='truckpad')

query = """SELECT *
             FROM trajetos t
             JOIN motoristas m ON (m.motoristas_id = t.motoristas_id)
             JOIN tipo_cnh c ON (c.tipo_cnh_id = m.tipo_cnh)
             JOIN tipo_caminhao tc ON (tc.tipo_caminhao_id = t.tipo_caminhao)"""
df = psql.read_sql(query, con=db)

db.close()
```
